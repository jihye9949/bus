```python
import numpy as np
import pandas as pd
import folium
from pandas import DataFrame   # 데이터 분석 패키지
from pandas import ExcelFile 
```

```python
bus = pd.read_csv('bus_train.csv', encoding = 'utf-8')
bus.head(5)
bus.tail(5)
```

```python
lat = bus['now_latitude'].mean()
long = bus['now_longitude'].mean()

map_osm = folium.Map([lat,long],zoom_start=9)
map_osm
```

```python
for i in range(1000):
    # 행 우선 접근 방식으로 값 추출하기
    name = bus.loc[i, 'now_station']
    lat = bus.loc[i, 'now_latitude']
    lng = bus.loc[i, 'now_longitude']
    
    # 추출한 정보를 지도에 표시
    marker = folium.Marker([lat,lng], popup=name)
    marker.add_to(map_osm)

map_osm.save('제주도 버스 정류장.html')
map_osm
```

```python
K_df = bus[['now_longitude', 'now_latitude']]
```

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters = 10, init='k-means++', max_iter=100)
kmeans.fit(K_df)
```
